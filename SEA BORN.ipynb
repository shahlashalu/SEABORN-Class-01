# **SEA BORN**

Seaborn is a Python data Visualization library built on top of Matplotlib.

It provides a high-level interface for creating aesthetically pleasing statical graphics.

Seaborn is particularly useful for exploring and visualizing datasets, allowing users to gain insights into their data quickly and effectively.

One of the key features of Seaborn is its ability to create visually appealing plots with minimal code.

It offers a wide range of plot types, including scatter plots, line plots, bar plots, histograms, box plots, violin plots, and more.

```python
import seaborn as sns
```

**SEABORN-DATASET**

Seaborn comes with several datasets that are built-in and can be directly accessed for visualization and analysis purposes.

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = df)
```

**Linear line is plotted according to data distribution.**

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = df,fit_reg = False)
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = df,fit_reg = False,hue = 'sex')
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = df,fit_reg = False,hue = 'sex',col = 'smoker')
```

```python
df['time'].value_counts()
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = df,fit_reg = False,hue = 'sex',col = 'smoker',row = 'time')
```

# **BOX PLOT**

A box plot (also known as a box-and-whisker plot) is a graphical representation of the distribution of a dataset based on five summary statistics : **minimum,first quartile(Q1),median(Q2),third quartile(Q3),and maximum.**

It is useful for detecting outliers and comparing distribution between different groups or categories.

```python
sns.boxplot(df['tip'])
```

```python
sns.boxplot(x = 'tip',y = 'day',data = df)
```

```python
sns.boxplot(x = 'tip',y = 'day',data = df,hue = 'smoker')
```

# **SWARM PLOT**

A swarm plot is similar to a strip plot, but it spreads out the points to avoid overlap, giving a better visualization of the distribution of the data .

Seaborn's swarmplot() function is used to create swarm plots.

Swarm plot is like the scatter plot where points donot overlap.


```python
sns.boxplot(x = 'day',y = 'total_bill',data = df)
sns.swarmplot(x= 'day',y = 'total_bill',data = df,color = ".10")
```
